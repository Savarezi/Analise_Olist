```python
df_pedidos.info()
```

```python
df_pedidos.shape
```

```python
#contando a quantidade de valores nulos
df_pedidos.isnull().sum()
```

```python
# Vamos começar nossa limpeza filtrando somente as colunas que desejamos utilizar e filtrando somente pedidos finalizados:

columns_drop = ['order_approved_at', 'order_delivered_carrier_date', 'order_estimated_delivery_date']
df_pedidos = df_pedidos.drop(columns_drop, axis=1)
df_pedidos
```

```python
# Agora iremos filtrar somente aqueles pedidos que foram finalizados, mas antes é importante entender todos os status disponíveis que temos na base:

df_pedidos['order_status'].unique()
```

```python
df_pedidos = df_pedidos.loc[df_pedidos['order_status'] == 'delivered']
df_pedidos
```

```python
# Filtrando somente pedidos finalizados, percebemos que perdemos um número razoável de linhas, cerca de 3 mil.
# Agora, com nosso dataset filtrado, precisamos resetar o seu index:
df_pedidos = df_pedidos.reset_index(drop=True)
df_pedidos
```

```python
# Para fecharmos a limpeza dessa tabela, precisamos criar uma coluna de mês! Uma das perguntas que precisamos responder é se existe sazonalidade
# nas vendas, então, criar uma coluna de mês irá facilitar bastante nosso trabalho na etapa de análise:

# Vamos transformar nossas colunas de data (que estão como string) em data:
df_pedidos['order_purchase_timestamp'] = pd.to_datetime(df_pedidos['order_purchase_timestamp'])
df_pedidos['order_delivered_customer_date'] = pd.to_datetime(df_pedidos['order_delivered_customer_date'])

# Agora sim, vamos criar uma coluna de mês!
df_pedidos['reference_month'] = df_pedidos['order_purchase_timestamp'].dt.strftime('%Y-%m-01')
df_pedidos
```

# Vamos começar o análise:

* Quantas vendas foram feitas no credito?
* Quantas vendas foram feitas no débito?

* Quantas vendas foram feitas a vista?

```python
# Filtrar as vendas feitas no credito
vendas_credito = df_olist[df_olist['payment_type'] == 'credit_card']

# Contar quantas vendas foram feitas no credito
quantidade_vendas_credito = vendas_credito.shape[0]
print(f'Quantidade de vendas feitas no crédito: {quantidade_vendas_credito}')
```

```python
# Filtrar as vendas feitas no débito
vendas_debito = df_olist[df_olist['payment_type'] == 'debit_card']

# Contar quantas vendas foram feitas no débito
quantidade_vendas_debito = vendas_debito.shape[0]
print(f'Quantidade de vendas feitas no débito: {quantidade_vendas_debito}')
```

```python
# Filtrar as vendas feitas no boleto
vendas_boleto = df_olist[df_olist['payment_type'] == 'boleto']

# Contar quantas vendas foram feitas no boleto
quantidade_vendas_boleto = vendas_boleto.shape[0]
print(f'Quantidade de vendas feitas no boleto: {quantidade_vendas_boleto}')
```

```python
# Filtrar as vendas feitas a vista
vendas_voucher = df_olist[df_olist['payment_type'] == 'voucher']

# Contar quantas vendas foram feitas a vista
quantidade_vendas_voucher = vendas_voucher.shape[0]
print(f'Quantidade de vendas feitas a vista: {quantidade_vendas_voucher}')
```

# Quais os 3 estados que mais compraram no debito

```python
# Identificar as 3 cidades que mais compraram no débito
cidades_mais_compraram = vendas_debito['customer_city'].value_counts().head(3)
print(f'As 3 cidades que mais compraram no débito:\n{cidades_mais_compraram}')
```

```python
# aqui estou mostrando o estado que mais compra, não importa a forma de pagamento

# Contar o número de compras por estado
compras_por_estado = df_olist['customer_state'].value_counts()

# Identificar o estado com o maior número de compras
estado_mais_compras = compras_por_estado.idxmax()
quantidade_mais_compras = compras_por_estado.max()

print(f'O estado que mais compra é {estado_mais_compras} com {quantidade_mais_compras} compras.')
```

Aqui está o gráfico de barras mostrando a quantidade de compras por estado. Esse tipo de gráfico permite uma comparação clara entre os estados, destacando São Paulo (SP) como o estado com o maior número de compras, seguido por Rio de Janeiro (RJ) e Minas Gerais (MG).

```python
import matplotlib.pyplot as plt

# Contando as compras por estado
compras_por_estado = df_olist['customer_state'].value_counts()

# Criando um gráfico de barras para os estados que mais compram
plt.figure(figsize=(10, 6))
compras_por_estado.plot(kind='bar', color='skyblue')
plt.title('Quantidade de Compras por Estado')
plt.xlabel('Estado')
plt.ylabel('Quantidade de Compras')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python


# Carregando os dados
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')

# Mesclando os dataframes
merged_df = pd.merge(orders, order_items, on='order_id')
merged_df = pd.merge(merged_df, products, on='product_id')

# Calculando o valor total de vendas por produto
sales_by_product = merged_df.groupby('product_id')['price'].sum()
print(sales_by_product)
```

# Identificando os Produtos Mais Vendidos:

```python


# Carregando os dados
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')

# Mesclando os dataframes
merged_df = pd.merge(orders, order_items, on='order_id')
merged_df = pd.merge(merged_df, products, on='product_id')

# Calculando o valor total de vendas por produto e quantidade vendida
sales_by_product = merged_df.groupby('product_id').agg({'price': 'sum', 'order_id': 'count'})
sales_by_product.columns = ['valor_total_vendas', 'quantidade_vendida']

# Ordenando os produtos pelo valor total de vendas (descendente)
top_products = sales_by_product.sort_values(by='valor_total_vendas', ascending=False)

# Exibindo os 10 produtos mais vendidos
print(top_products.head(10))
```

 # Os 10 produtos mais vendidos.

```python
import matplotlib.pyplot as plt

# ... (código anterior para calcular as vendas por produto)

# Criando um gráfico de barras invertidas
top_products.head(10).plot(kind='barh', figsize=(10, 5))
plt.title('10 Produtos Mais Vendidos',color='green')
plt.xlabel('Valor Total de Vendas',color='green')
plt.ylabel('ID do Produto',color='green')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Exibindo o gráfico
plt.tight_layout

plt.show()
```
